```
using System;

// Интерфейс для научной работы
public interface IResearcher
{
    void ConductResearch();
    void PublishPaper();
}

// Интерфейс для навыков преподавания
public interface ITeachingAssistant
{
    void GradeAssignments();
    void ConductLabSessions();
}

// Базовый класс Student
public class Student
{
    private string name;
    private int age;
    private int course;

    public string Name
    {
        get => name;
        set => name = value;
    }

    public int Age
    {
        get => age;
        set => age = value;
    }

    public int Course
    {
        get => course;
        set => course = value;
    }

    public Student(string name, int age, int course)
    {
        Name = name;
        Age = age;
        Course = course;
    }

    public virtual string GetInfo()
    {
        return $"Студент: {Name}, Возраст: {Age}, Курс: {Course}";
    }

    public virtual void Study()
    {
        Console.WriteLine($"{Name} учится.");
    }

    public virtual void TakeExam()
    {
        Console.WriteLine($"{Name} сдаёт экзамен.");
    }

    public void Help(Student student)
    {
        Console.WriteLine($"{Name} помогает {student.Name} с учёбой.");
    }
}

// Производный класс для студентов бакалавриата
public class BachelorStudent : Student
{
    private string specialty;
    private double gpa;

    public string Specialty
    {
        get => specialty;
        set => specialty = value;
    }

    public double GPA
    {
        get => gpa;
        set => gpa = value;
    }

    public BachelorStudent(string name, int age, int course, string specialty, double gpa)
        : base(name, age, course)
    {
        Specialty = specialty;
        GPA = gpa;
    }

    public override string GetInfo()
    {
        return base.GetInfo() + $", Специальность: {Specialty}, Средний балл: {GPA}";
    }

    public override void Study()
    {
        Console.WriteLine($"{Name} учится по специальности {Specialty}.");
    }

    public void ApplyForInternship()
    {
        Console.WriteLine($"{Name} подаёт заявку на стажировку.");
    }
}

// Производный класс для студентов магистратуры, использует множественное наследование через интерфейсы
public class MasterStudent : Student, IResearcher, ITeachingAssistant
{
    private string thesisTitle;
    private string scientificAdvisor;

    public string ThesisTitle
    {
        get => thesisTitle;
        set => thesisTitle = value;
    }

    public string ScientificAdvisor
    {
        get => scientificAdvisor;
        set => scientificAdvisor = value;
    }

    public MasterStudent(string name, int age, int course, string thesisTitle, string scientificAdvisor)
        : base(name, age, course)
    {
        ThesisTitle = thesisTitle;
        ScientificAdvisor = scientificAdvisor;
    }

    public override string GetInfo()
    {
        return base.GetInfo() + $", Тема диссертации: {ThesisTitle}, Научный руководитель: {ScientificAdvisor}";
    }

    public override void TakeExam()
    {
        Console.WriteLine($"{Name} сдаёт экзамен под руководством {ScientificAdvisor}.");
    }

    // Реализация интерфейса IResearcher
    public void ConductResearch()
    {
        Console.WriteLine($"{Name} проводит исследование по теме {ThesisTitle}.");
    }

    public void PublishPaper()
    {
        Console.WriteLine($"{Name} публикует статью по своей диссертации.");
    }

    // Реализация интерфейса ITeachingAssistant
    public void GradeAssignments()
    {
        Console.WriteLine($"{Name} проверяет задания студентов.");
    }

    public void ConductLabSessions()
    {
        Console.WriteLine($"{Name} проводит лабораторные занятия.");
    }
}

// Производный класс для аспирантов, использует множественное наследование через интерфейсы
public class PhDStudent : Student, IResearcher
{
    private string dissertationTopic;
    private int publicationsCount;

    public string DissertationTopic
    {
        get => dissertationTopic;
        set => dissertationTopic = value;
    }

    public int PublicationsCount
    {
        get => publicationsCount;
        set => publicationsCount = value;
    }

    public PhDStudent(string name, int age, int course, string dissertationTopic, int publicationsCount)
        : base(name, age, course)
    {
        DissertationTopic = dissertationTopic;
        PublicationsCount = publicationsCount;
    }

    public override string GetInfo()
    {
        return base.GetInfo() + $", Тема диссертации: {DissertationTopic}, Количество публикаций: {PublicationsCount}";
    }

    public override void Study()
    {
        Console.WriteLine($"{Name} учится и проводит исследования по теме {DissertationTopic}.");
    }

    // Реализация интерфейса IResearcher
    public void ConductResearch()
    {
        Console.WriteLine($"{Name} проводит научные исследования по теме: {DissertationTopic}.");
    }

    public void PublishPaper()
    {
        Console.WriteLine($"{Name} публикует статью по своему исследованию.");
    }

    public void ApplyForGrant()
    {
        Console.WriteLine($"{Name} подаёт заявку на грант для финансирования исследования.");
    }
}

public class Scientist : PhDStudent
{
    public int Experience {get;set;}
    public Scientist(string name, int age, int course, string dissertationTopic, int publicationsCount, int experience) : base (name, age, course, dissertationTopic, publicationsCount)
    {
        Experience = experience;
    }

    public void DisplayWorkYears()
    {
        Console.WriteLine($"Опыт работы: {Experience}");
    }
}


// Пример использования классов

BachelorStudent bachelor = new BachelorStudent("Иван", 20, 2, "Информатика", 4.5);
MasterStudent master = new MasterStudent("Мария", 24, 1, "Компьютерное зрение", "Проф. Сидоров");
PhDStudent phd = new PhDStudent("Алексей", 28, 3, "Машинное обучение", 5);
Scientist scn = new Scientist("Алексей", 28, 3, "Машинное обучение", 5, 5);


Console.WriteLine(bachelor.GetInfo());
Console.WriteLine(master.GetInfo());
Console.WriteLine(phd.GetInfo());

// Демонстрация полиморфизма
bachelor.Study();
master.TakeExam();
phd.Study();

// Взаимодействие объектов
bachelor.Help(master);
master.Help(phd);

// Используем методы интерфейсов
master.ConductResearch();
master.GradeAssignments();
phd.PublishPaper();
phd.ApplyForGrant();

scn.DisplayWorkYears();
```
