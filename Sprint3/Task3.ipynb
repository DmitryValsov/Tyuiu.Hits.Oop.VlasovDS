Создать базовый класс Student в C#, который будет представлять информацию о
студентах. На основе этого класса разработать 2-3 производных класса,
демонстрирующих принципы наследования и полиморфизма. В каждом из классов
должны быть реализованы новые атрибуты и методы, а также переопределены
некоторые методы базового класса для демонстрации полиморфизма.

#### Дополнительное задание
Добавьте к сущестующим классам (базовыму и производным 3-4 атрибута и метода) и реализуйте полиморфизм с перекрытием и прегегрузкой методов, а также generic классы

```
using System;

// Интерфейс для научной работы
public interface IResearcher
{
    void ConductResearch();
    void PublishPaper();
}

// Интерфейс для навыков преподавания
public interface ITeachingAssistant
{
    void GradeAssignments();
    void ConductLabSessions();
}

// Базовый класс Student
public class Student
{
    private string name;
    private int age;
    private int course;

    public string Name
    {
        get => name;
        set => name = value;
    }

    public int Age
    {
        get => age;
        set => age = value;
    }

    public int Course
    {
        get => course;
        set => course = value;
    }

    public Student(string name, int age, int course)
    {
        Name = name;
        Age = age;
        Course = course;
    }

    public virtual string GetInfo()
    {
        return $"Студент: {Name}, Возраст: {Age}, Курс: {Course}";
    }

    public virtual void Study()
    {
        Console.WriteLine($"{Name} учится.");
    }

    public virtual void TakeExam()
    {
        Console.WriteLine($"{Name} сдаёт экзамен.");
    }

    public void Help(Student student)
    {
        Console.WriteLine($"{Name} помогает {student.Name} с учёбой.");
    }
}

// Производный класс для студентов бакалавриата
public class BachelorStudent : Student
{
    private string specialty;
    private double gpa;

    public string Specialty
    {
        get => specialty;
        set => specialty = value;
    }

    public double GPA
    {
        get => gpa;
        set => gpa = value;
    }

    public BachelorStudent(string name, int age, int course, string specialty, double gpa)
        : base(name, age, course)
    {
        Specialty = specialty;
        GPA = gpa;
    }

    public override string GetInfo()
    {
        return base.GetInfo() + $", Специальность: {Specialty}, Средний балл: {GPA}";
    }

    public override void Study()
    {
        Console.WriteLine($"{Name} учится по специальности {Specialty}.");
    }

    public void ApplyForInternship()
    {
        Console.WriteLine($"{Name} подаёт заявку на стажировку.");
    }

    // Перегрузка метода
    public void ApplyForInternship(string company)
    {
        Console.WriteLine($"{Name} подаёт заявку на стажировку в компанию {company}.");
    }
}

// Производный класс для студентов магистратуры, использует множественное наследование через интерфейсы
public class MasterStudent : Student, IResearcher, ITeachingAssistant
{
    private string thesisTitle;
    private string scientificAdvisor;
    public string ResearchArea { get; set; } // Дополнительный атрибут

    public string ThesisTitle
    {
        get => thesisTitle;
        set => thesisTitle = value;
    }

    public string ScientificAdvisor
    {
        get => scientificAdvisor;
        set => scientificAdvisor = value;
    }

    public MasterStudent(string name, int age, int course, string thesisTitle, string scientificAdvisor, string researchArea)
        : base(name, age, course)
    {
        ThesisTitle = thesisTitle;
        ScientificAdvisor = scientificAdvisor;
        ResearchArea = researchArea;
    }

    public override string GetInfo()
    {
        return base.GetInfo() + $", Тема диссертации: {ThesisTitle}, Научный руководитель: {ScientificAdvisor}, Область исследований: {ResearchArea}";
    }

    public override void TakeExam()
    {
        Console.WriteLine($"{Name} сдаёт экзамен под руководством {ScientificAdvisor}.");
    }

    // Реализация интерфейса IResearcher
    public void ConductResearch()
    {
        Console.WriteLine($"{Name} проводит исследование по теме {ThesisTitle} в области {ResearchArea}.");
    }

    public void PublishPaper()
    {
        Console.WriteLine($"{Name} публикует статью по своей диссертации.");
    }

    // Реализация интерфейса ITeachingAssistant
    public void GradeAssignments()
    {
        Console.WriteLine($"{Name} проверяет задания студентов.");
    }

    public void ConductLabSessions()
    {
        Console.WriteLine($"{Name} проводит лабораторные занятия.");
    }
}

// Generic-класс для работы с различными студентами
public class University<T> where T : Student
{
    private List<T> students = new List<T>();

    public void AddStudent(T student)
    {
        students.Add(student);
        Console.WriteLine($"Добавлен студент: {student.Name}");
    }

    public void ShowAllStudents()
    {
        foreach (var student in students)
        {
            Console.WriteLine(student.GetInfo());
        }
    }
}


        // Демонстрация базового класса
        Student student = new Student("Иван", 18, 1);
        Console.WriteLine(student.GetInfo());
        student.Study();

        // Демонстрация производных классов
        BachelorStudent bachelor = new BachelorStudent("Анна", 20, 3, "Программирование", 4.5);
        Console.WriteLine(bachelor.GetInfo());
        bachelor.ApplyForInternship("Google");

        MasterStudent master = new MasterStudent("Елена", 24, 2, "ИИ в медицине", "Др. Смит", "Машинное обучение");
        Console.WriteLine(master.GetInfo());
        master.ConductResearch();

        // Использование generic-класса
        var university = new University<Student>();
        university.AddStudent(student);
        university.AddStudent(bachelor);
        university.AddStudent(master);

        Console.WriteLine("\nВсе студенты:");
        university.ShowAllStudents();
  
```
